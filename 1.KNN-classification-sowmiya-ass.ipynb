```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Load datset
dataset=pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
#Convert categorical variables to dummy variables
dataset=pd.get_dummies(dataset,drop_first=True)
dataset=dataset.astype(int)
dataset
```

```python
print(dataset["classification_yes"].value_counts())
```

```python
# split dataset into independent and dependent variables
indep=dataset.drop("classification_yes", axis=1)
```

```python
dep=dataset["classification_yes"]
```

```python
indep.shape
```

```python
dep
```

```python
#split dataset into training and test sets
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(indep,dep,test_size=1/3,random_state=0)
```

```python
#initialize and train KNNClassifier
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors =7, metric='minkowski', p =2)
classifier.fit(X_train,y_train)
```

```python
#Evaluate the model
y_pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
```

```python
print(cm)
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(y_test,y_pred)
```

```python
print(clf_report)
```

```python
#predict using the trained model 
sample_data=[[40,70,1,0,100,30,1,140,4,15,False,False,False,True,True,False,True,False,True]]
prediction=classifier.predict(sample_data)
```
